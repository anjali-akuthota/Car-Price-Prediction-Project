```python
import pandas as pd
d=pd.read_csv('/content/CAR DETAILS FROM CAR DEKHO.csv')
print(d)
```

```python
y=d['selling_price']
print(y)
```

```python
x=d.drop(columns=['selling_price'])
print(x)
x=x.drop(columns=['name'])
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(x.shape)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=True,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
mm=lr.fit(x_train,y_train)
```

```python
yp=mm.predict(x_test)
```

```python
print(yp)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(yp,y_test))
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(yp,y_test))
```

```python
mse = mean_squared_error(y_test, yp)
print("Mean Squared Error:", mse)
```

```python
mae = mean_absolute_error(y_test, yp)
print("Mean Absolute Error:", mae)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
